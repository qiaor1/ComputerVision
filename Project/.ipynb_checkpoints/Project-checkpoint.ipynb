## Milestone2: Modify reconstruct function to make use of the two-color images collected from each camera

```python
def color_decode(pts2L,pts2R,imprefix,start,threshold):
    """
    Given a sequence of 20 images of a scene showing projected 10 bit gray code, 
    decode the binary sequence into a decimal value in (0,1023) for each pixel.
    Mark those pixels whose code is likely to be incorrect based on the user 
    provided threshold.  Images are assumed to be named "imageprefixN.png" where
    N is a 2 digit index (e.g., "img00.png,img01.png,img02.png...")
 
    Parameters
    ----------
    imprefix : str
       Image name prefix
      
    start : int
       Starting index
       
    threshold : float
       Threshold to determine if a bit is decodeable
       
    Returns
    -------
    code : 2D numpy.array (dtype=float)
        Array the same size as input images with entries in (0..1023)
        
    mask : 2D numpy.array (dtype=logical)
        Array indicating which pixels were correctly decoded based on the threshold
    
    """
    
    # we will assume a 10 bit code
    nbits = 10
    gray=[]
    mask_t=[]
    colorL = plt.imread(imprefix+"C0_00.png")
    colorFL= plt.imread(imprefix+"C0_01.png")
    colorR= plt.imread(imprefix+"C1_00.png")
    colorFR= plt.imread(imprefix+"C1_01.png")
    bvaluesL_list=[]
    bvaluesR_list=[]
    for i in range(pts2L.shape[1]):
        bvaluesL_list.append(colorL[pts2L[1][i]][pts2L[0][i]])
        bvaluesR_list.append(colorR[pts2R[1][i]][pts2R[0][i]])
    bvaluesL=np.array(bvaluesL_list).T
    bvaluesR=np.array(bvaluesR_list).T
    bvalues=(bvaluesL+bvaluesR)/2
    colorL= colorL-colorFL
    colorR= colorR-colorFR

    print(colorL.shape)
    print(colorR.shape)
    mask_t= np.asarray(mask_t)
    mask_t=mask_t.astype(int)
    maskL= np.zeros((colorL.shape[0],colorL.shape[1]))
    maskR= np.zeros((colorR.shape[0],colorR.shape[1]))
    undecodableL=np.where(np.abs(colorL) >threshold )
    undecodableR=np.where(np.abs(colorR) >threshold )
    for i in range(undecodableL[1].size):
        maskL[undecodableL[0][i]][undecodableL[1][i]]=1
    for i in range(undecodableR[1].size):
        maskR[undecodableR[0][i]][undecodableR[1][i]]=1
    color= (colorL+colorR)/2
    c=color.reshape((-1,3)).T
    
    
    return bvalues,maskL,maskR
```

```python
## Mesh cleanup parameters
from scipy.spatial import Delaunay
# Specify limits along the x,y and z axis of a box containing the object
# we will prune out triangulated points outside these limits
# given x,y,z [-200 400],[-200 300], and [-200 200]
boxlimits = np.array([0,19,-16,10,27,40])

# Specify a longest allowed edge that can appear in the mesh. Remove triangles
# from the final mesh that have edges longer than this value
trithresh = 30

print(pts3.shape)
print(pts2L.shape)
print(pts2R.shape)
#
# bounding box pruning
#
drop=[]
for i in range(pts3.shape[1]):
    if (pts3[0][i]<boxlimits[0] or pts3[0][i]>boxlimits[1] or 
        pts3[1][i]<boxlimits[2] or pts3[1][i]>boxlimits[3] or 
        pts3[2][i]<boxlimits[4] or pts3[2][i]>boxlimits[5]):
        drop.append(i)
        
pts3=np.delete(pts3, drop, axis=1) 
pts2L=np.delete(pts2L, drop, axis=1) 
pts2R=np.delete(pts2R, drop, axis=1)
color=np.delete(color, drop, axis=1)

print(pts3.shape)
print(pts2L.shape)
print(pts2R.shape)


#
# triangulate the 2D points to get the surface mesh
#

tri= Delaunay(pts2L.T).simplices

#
# triangle pruning
#
def dis(a,b,c,tri): 
    return np.sqrt(np.sum(np.power((pts3.T[tri[a][b]]-pts3.T[tri[a][c]]),2), axis=0))
# drop_pts=[] 
drop_tri=[] 
for i in range(tri.shape[0]):
    if (dis(i,0,1,tri)>trithresh 
    or dis(i,0,2,tri)>trithresh or dis(i,1,2,tri)>trithresh): 
        drop_tri.append(i)
        
tri=np.delete(tri,drop_tri,axis=0)



```

```python
tokeep=np.unique(tri)
map = np.zeros(pts3.shape[1])
pts3=pts3[:,tokeep]
color = color[:,tokeep]

#update tri
map[tokeep] = np.arange(0,tokeep.shape[0])
tri=map[tri]
```
